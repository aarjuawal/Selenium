```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
```

```python
def setup():
    driver = webdriver.Chrome()
    driver.get("https://quotes.toscrape.com/")
    return driver
```

```python
def teardown(driver):
    driver.quit()
def get_quotes():
    driver = setup()
    pTag = driver.find_elements(By.XPATH, '//div[@class="quote"]/span[@itemprop="text"]')
    for val in pTag:
        print(val.text)
    teardown(driver)
get_quotes()
```

```python
def get_author():
    driver = setup()
    pTag = driver.find_elements(By.XPATH, '//small[@class="author"]')
    for val in pTag:
        print(val.text)
    teardown(driver)
get_author()
```

```python
def get_tags():
    driver = setup()
    pTag = driver.find_elements(By.XPATH, '//div[@class="tags"]/a')
    for val in pTag:
        print(val.text)
    teardown(driver)
get_tags()
```

```python
def GetButton():
    driver = setup()
    nxtBtn = driver.find_element(By.XPATH, '//li[@class="next"]/a')
    nxtBtn.click()
    teardown(driver)
GetButton()
```

```python
def love_btn():
  driver=setup()
  loveBtn =driver.find_element(By.XPATH, '//span[@class="tag-item"]/a[@class="tag" and text()="love"]')
  loveBtn.click()
  try:
    actres=driver.find_element(By.XPATH, "//div[@class='container']/h3")
    actualResult = actres.text
    expectedResult= "Viewing tag: love"
    assert actualResult == expectedResult, 'Failed'
    print("Success!")
  except AssertionError as e:
    print(e)
  teardown(driver)
love_btn()
```

```python
def scrolldown(driver):

    page_height = driver.execute_script("return document.body.scrollHeight")
    scroll_speed = 1
    scroll_iteration = int(page_height / scroll_speed)
    for _ in range(scroll_iteration):
        driver.execute_script(f"window.scrollBy(0,{scroll_speed});")
    time.sleep(2)
```

```python
def Login():
    driver = setup()
    LoginBtn = driver.find_element(By.XPATH, '//a[@href="/login"]')
    LoginBtn.click()
    username= driver.find_element(By.ID, "username")
    username.send_keys("aa@gmail.com")
    time.sleep(2)
    password= driver.find_element(By.ID, "password")
    password.send_keys("akkaka")
    time.sleep(2)
    submit = driver.find_element(By.XPATH, "//input[@type='submit']")
    submit.click()
    time.sleep(3)
    scrolldown(driver)
    time.sleep(2)
    GetButton()
    teardown(driver)
Login()
```

```python
def get_quote():
    driver = setup()
    quotes = driver.find_elements(By.CLASS_NAME, 'text')
    authors = driver.find_elements(By.CLASS_NAME, 'author')
    tags_list = driver.find_elements(By.CLASS_NAME, 'tags')

    for i in range(len(quotes)):
        print(f"'{quotes[i].text}'")
        print(f"-by{authors[i].text}")
        tags = tags_list[i].find_elements(By.TAG_NAME,'a')
        tag_texts = [tag.text for tag in tags]
        if tag_texts:
            print("Tags:", ", ".join(tag_texts))
        else:
            print("Tags: None")

    teardown(driver)
get_quote() 
```
